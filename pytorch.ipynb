```
    import numpy as np
    from tensorflow.python.keras import backend

    # 随机输出数字0~10的概率分布
    output = backend.random_uniform_variable(shape=(1, 10), low=0, high=1)
    # 实际结果假设为数字1
    actual_pos = backend.variable(np.array([1]), dtype='int32')
    print("数字0~9的预测概率分布为:", backend.eval(output))
    print("实际结果为数字:", backend.eval(actual_pos))
    print("实际结果是否in top 1: ", backend.eval(backend.in_top_k(output, actual_pos, 1)))
    print("实际结果是否in top 5: ", backend.eval(backend.in_top_k(output, actual_pos, 5)))
```

```
import torch
import torch.nn as nn
import math

# 多分类问题，交叉熵计算
entroy=nn.CrossEntropyLoss()
input=torch.Tensor([[-0.7715, -0.6205,-0.2562]])
target = torch.tensor([1])
output = entroy(input, target)
print(output)

# 手动验证
input = [-0.7715, -0.6205,-0.2562]
res0 = -input[1] + math.log(sum([math.exp(i) for i in input]))
print(res0)
```
